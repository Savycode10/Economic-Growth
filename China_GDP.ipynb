```python
import pandas as pd
df = pd.read_excel('China GDP.xlsx')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Set the style for the plot
sns.set_theme(style="whitegrid")
# Plot the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='Year', y='GDP (current US$)', marker='o')
plt.title('Real GDP of China', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (current US$)', fontsize=12)
plt.grid(True)
plt.show()
```

```python

import numpy as np

# Set the style for the plot
sns.set_theme(style="whitegrid")

plt.figure(figsize=(12, 7))

# Plot the time series for GDP growth (annual %)
sns.lineplot(data=df, x='Year', y='GDP growth (annual %)', marker='o', label='GDP Growth (Annual %)')

# Fit a linear regression model
z = np.polyfit(df['Year'], df['GDP growth (annual %)'], 1)
p = np.poly1d(z)

# Add the linear growth line to the plot
plt.plot(df['Year'], p(df['Year']), color='red', linestyle='--', label=f'Linear Fit: y = {z[0]:.2f}x + {z[1]:.2f}')

plt.title('GDP Growth (Annual %) of China', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP Growth (Annual %)', fontsize=12)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import pandas as pd

# Define custom bins and labels
bins = [1961, 1971, 1981, 1991, 2001, 2011, 2021, 2025]  # Note: last bin ends at 2025
labels = [
    '1961-1970',
    '1971-1980',
    '1981-1990',
    '1991-2000',
    '2001-2010',
    '2011-2020',
    '2021-2024'
]

# Assign each year to its corresponding interval
df['Decade'] = pd.cut(df['Year'], bins=bins, labels=labels, right=False)

# Calculate average GDP growth by decade
decade_avg = df.groupby('Decade')['GDP growth (annual %)'].mean().reset_index()

# Round for neatness
decade_avg['GDP growth (annual %)'] = decade_avg['GDP growth (annual %)'].round(2)

# Export to CSV
decade_avg.to_csv('china_gdp_growth_by_decade.csv', index=False)

# Print to confirm
print("Exported successfully with final range as 2021–2024:\n")
print(decade_avg)
```
